```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# label count折線圖
def label_plot(df):
  plt.figure(figsize=(12, 6))
  value_counts = df['label'].value_counts()

  plt.xticks(rotation=45)  

  # 繪製折線圖
  plt.plot(value_counts.index, value_counts.values, marker='o', linestyle='-')

  # 加上標題和標籤
  plt.title('Line Chart of label Counts')
  plt.xlabel('label')
  plt.ylabel('Count')

  # 調整圖表大小
  plt.tight_layout()

  # 顯示折線圖
  plt.show()


def merge_content(row):
    if pd.notna(row['time_normalized']):
        row['content'] = row['content'] + '=>' + row['time_normalized']
    return row['content']

def txt_2_df(path):
  df = pd.read_csv(path, sep='\t',header=None,names=['file_name', 'label', 'start', 'end', 'content', 'time_normalized'])

  df['content'] = df.apply(merge_content, axis=1)
  df = df.drop(columns=['time_normalized'])
  return df

def get_sub_df(df, label):
  return df[df['label'] == label]
```

# 合併df
## 加入GPT生成的資料

```python
origin_file_path = "/content/drive/MyDrive/dataset/merge_df.tsv"

# 使用 pandas 載入 TSV 檔案
df = pd.read_csv(origin_file_path, delimiter='\t', names=['file_name', 'start', 'content', 'answer'])

# 顯示資料
df
```

```python
train_file_name = df['file_name'].unique()
train_file_name_len = df['file_name'].nunique()

# 將chatGPT生成的單筆資料轉成訓練格式的dataframe
def txt_to_df(file_path):
  file_name = []
  start = []
  content = []
  answer = []

  with open(file_path, 'r', encoding='utf-8') as file:
    contents = file.readlines()

  start1 = 0
  end1 = train_file_name_len - 1

  start2 = 500
  end2 = 2500

  for c in contents:
    if c != '\n':
      file_name.append(train_file_name[random.randint(start1, end1)])
      start.append(random.randint(start2, end2))
      content.append(c.split('\\t')[0])
      answer.append(c.split('\\t')[1].split('\n')[0])


  content_dict = {"file_name": file_name, "start": start, "content": content, "answer": answer}
  processed_df = pd.DataFrame(content_dict)
  return processed_df
```

```python
# 將每筆chatGPT生成資料的dataframe跟原有資料(根據檔名)合併
def concat_df(origin_df, sub_df):
  new_df = pd.DataFrame({"file_name":[], "start":[], "content":[], "answer":[]})

  sub_list = sub_df['file_name'].unique()

  for f in train_file_name:
    if f in sub_list:
      a = origin_df[origin_df['file_name'] == f]
      b = sub_df[sub_df['file_name'] == f]
      result_df = pd.concat([a, b], ignore_index=True)
      new_df = pd.concat([new_df, result_df], ignore_index=True)
    else:
      origin_sub_df = origin_df[origin_df['file_name'] == f]
      new_df = pd.concat([new_df, origin_sub_df], ignore_index=True)

  new_df['start'] = new_df['start'].astype(int)

  return new_df
```
